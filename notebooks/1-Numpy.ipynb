## 1.2 Operaciones con vectores
Python tiene por defecto distintas estructuras de almacenamiento de datos, pero cuando necesitamos hacer operaciones matemáticas las listas o tuplas no nos sirven. Necesitamos trabajar con vectores o matrices, lo que se suele llamar **N-arrays** (formaciones de N dimensiones). Vemos un ejemplo de lo que hace la suma de dos listas:

```python
l1 = [1, 3, 4]
l2 = [1, 4, 5]
print(type(l1),type(l2))
print(f'la suma de {l1} y {l2} es igual a {l1 + l2}')
```

Con numpy podemos generar *arrays* que sí que permiten la operación suma que queremos hacer:

```python
a1 = np.array([1,3,4])
a2 = np.array(l2)
print(type(a1),type(a2))
print(f'la suma de {a1} y {a2} es igual a {a1 + a2}')
```

```python
a1 * 2       # multiplica todos los elementos * 2
```

```python
# en cambio la lista multiplicada por 2 es: 
l1 * 2
```

```python
# también podemos multiplicar elemento a elemento
a1 * a2  
```

```python
# comprobar  si umple alguna condición ...
a1 > 1
```

## 1.5 Números aleatorios
**NumPy** nos permite generar número aleatorios con distintas distribuciones. Aquí mostraremos las más comunes pero podéis ver todas las que tiene definidas [aquí](https://numpy.org/doc/1.16/reference/routines.random.html).

```python

# Distribución normal, le damos: media, dispersión y tamaño de la muestra.
r1 = np.random.normal(1, 2, 1000)  
r1
```

```python
import matplotlib.pyplot as plt
plt.hist(r1)
```

```python
# Distribución uniforme, le damos mínimo, máximo y tamaño muestra. 
r2 = np.random.uniform(1, 2, 1000) 
r2
```

```python
plt.hist(r2)
```

## 1.7 Ajustes
Numpy nos permite hacer ajustes a polinomios. Para ajustes a otro tipo de funciones podemos usar `scipy`. 

Muchas veces tenemos unos datos y nos interesa sacar un modelo. Cuando no tenemos un modelo teórico podemos ajustar un polinomio. Si el plonomio es de base 1, es un ajuste lineal
$y = ax + b$, durante el ajuste obtenemos los parámetros de la recta a y b.

```python
x = np.arange(10)
y = x * 2 + np.random.normal(2, size=10)

plt.plot(x, y, '*')
plt.show()
```

```python
a,b = np.polyfit(x,y,1)
a,b
```

```python
plt.plot(x, y, '*', label = 'data')
plt.plot(x, a*x + b, label = 'fit')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# o más compacta
plt.plot(x ,y,'*')
par = np.polyfit(x,y,1)
obj1 = np.poly1d(par)     # guardas el polinomio dados unos coeficientes
plt.plot(x, obj1(x))       # 
```

```python
obj1(x)
```
